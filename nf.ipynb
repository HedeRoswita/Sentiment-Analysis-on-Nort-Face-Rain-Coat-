```python
cookie={} # insert request cookies within{}
def getAmazonSearch(search_query):
    url="https://www.amazon.com/s?k="+search_query
    print(url)
    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchasin(asin):
    url="https://www.amazon.com/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):
    url="https://www.amazon.com"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
product_names=[]
response=getAmazonSearch('north+face+rain+coat')
soup=BeautifulSoup(response.content)

for i in soup.findAll("span",{'class':'a-size-base-plus a-color-base a-text-normal'}): # the tag which is common for all the names of products
    product_names.append(i.text) #adding the product names to the list
```

```python
data_asin=[]
response=getAmazonSearch('north+face+rain+coat')
soup=BeautifulSoup(response.content)
for i in soup.findAll("div",{'class':'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20'}):
    data_asin.append(i['data-asin'])
```

```python
response.status_code
```

```python
link = []

for i in range(len(data_asin)):
    response = Searchasin(data_asin[i])

    # Check if response is valid
    if isinstance(response, str):
        print(f"Error: {response}")
        continue

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all review links and append them to the link list
    for anchor in soup.findAll("a", {'data-hook': "see-all-reviews-link-foot"}):
        link.append(anchor['href'])
```

```python
review = []
star = []
date= []
user=[]
summary=[]
for j in range(len(link)):
    for k in range(5):
      response = Searchreviews(link[j] + '&pageNumber=' + str(k))  # Get response for current page
      soup = BeautifulSoup(response.content)
      review_containers = soup.findAll('div',{'class':'a-section review aok-relative'})
      for a in review_containers:
        review_span = i.find("span", {'data-hook': "review-body"})
        review.append(review_span.text)
      for b in review_containers:
        star_span = b.find("span", {'class': "a-icon-alt"})
        star.append(star_span)
      for c in review_containers:
        date_span = c.find("span", {'data-hook': "review-date"})
        date.append(date_span)
      for d in review_containers:
        user_span = d.find("span", {'class': 'a-profile-name'})
        user.append(user_span)
      for e in review_containers:
        summary_span = e.find("a", {'data-hook':"review-title"})
        summary.append(summary_span)




```

```python
#this code is working finee!!!!
review = []
star = []
date= []
user=[]
summary=[]
for j in range(len(link)):
    for k in range(60):
      response = Searchreviews(link[j] + '&pageNumber=' + str(k))  # Get response for current page
      soup = BeautifulSoup(response.content)
      review_containers = soup.findAll('div',{'class':'a-section review aok-relative'})
      for a in review_containers:
        review_span = a.find("span", {'data-hook': "review-body"})
        review_text = review_span.text.strip() if review_span else 'N/A'
        review.append(review_text)
      for b in review_containers:
        star_span = b.find("span", {'class': "a-icon-alt"})
        star_text = star_span.text.strip() if review_span else 'N/A'
        star.append(star_text)
      for c in review_containers:
        date_span = c.find("span", {'data-hook': "review-date"})
        date_text = date_span.text.strip() if review_span else 'N/A'
        date.append(date_text)
      for d in review_containers:
        user_span = d.find("span", {'class': 'a-profile-name'})
        user_text = user_span.text.strip() if review_span else 'N/A'
        user.append(user_text)
      for e in review_containers:
        summary_span = e.find("a", {'data-hook':"review-title"})
        summary.append(summary_span)

```

```python
pd.set_option('max_colwidth',800)
df6= pd.DataFrame({'user':user, 'date': date, 'review': review, 'summary': summary,'star': star})
df6
```
